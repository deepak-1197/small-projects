```
x=sample_df[['age','sex','cp','trestbps','chol','fbs','restecg','thalach','exang','oldpeak','slope','ca','thal']]
y=sample_df['target']
```

```
x
```

```
from sklearn.model_selection import train_test_split
```

```
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.3,random_state=10)
```

```
import seaborn as sns
sns.set_style("whitegrid")

sns.FacetGrid(sample_df,hue="target",size=4).map(plt.scatter,'chol','thalach',alpha=0.5).add_legend()

plt.show()

sns.FacetGrid(sample_df,hue="target",size=4).map(plt.scatter,'age','thalach',alpha=0.5).add_legend()

plt.show()

sns.FacetGrid(sample_df,hue="target",size=4).map(plt.scatter,'trestbps','thalach',alpha=0.5).add_legend()

plt.show()
```

```
len(x_train)
```

```
x_train
```

```
from sklearn import tree
clf=tree.DecisionTreeClassifier()

clf.fit(x_train,y_train)

```

```
clf.score(x_test,y_test)
```

```
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(n_estimators=50)
model.fit(x_train,y_train)
```

```
model.score(x_test,y_test)
```

```
from sklearn import svm
model2=svm.SVC(kernel='linear',gamma='auto')
model2.fit(x_train,y_train)
y_predict=model2.predict(x_test)
```

```
from sklearn.metrics import classification_report
classification_report(y_test,y_predict)
```

```
model2.score(x_test,y_test)
```

```
from sklearn.neighbors import KNeighborsClassifier
nn=KNeighborsClassifier(n_neighbors=3)
nn.fit(x_train,y_train)
predicted=nn.predict(x_test)

```

```
nn.score(x_test,y_test)
```
