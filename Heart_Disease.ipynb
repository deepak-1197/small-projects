```
sample_df=pd.read_csv(io.StringIO(uploaded['dataset1.csv'].decode('utf-8')))
```

```
import matplotlib.pyplot as plt
%matplotlib inline
```

```
x=sample_df[['age','sex','cp','trestbps','chol','fbs','restecg','thalach','exang','oldpeak','slope','ca','thal']]
y=sample_df['target']
```

```
x
```

```
from sklearn.model_selection import train_test_split
```

```
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.5,random_state=10)
```

```
len(x_train)
```

```
x_train
```

```
from sklearn import tree
clf=tree.DecisionTreeClassifier()

clf.fit(x_train,y_train)

```

```
clf.predict(x_test)
```

```
clf.predict([['58','0','0','130','197','0','1','131','0','0.6','1','0','2']])
```

```
y_test
```

```
clf.score(x_test,y_test)
```

```
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,clf.predict(x_test))
```

```
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
Log_ROC_auc=roc_auc_score(y_test,clf.predict(x_test))
fpr,tpr,thresholds=roc_curve(y_test,clf.predict_proba(x_test)[:,1])
```

```
y_test
```

```
plt.figure()
plt.plot(fpr,tpr, label="Clf 1(area = %.2f)" % Log_ROC_auc)
plt.show()
```
